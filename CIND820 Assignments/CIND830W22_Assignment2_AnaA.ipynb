**b)**  **[5 pts]**  Write a function to receive the Supermarket Sales list (that you created in the previous question) and return the total sales (the sum of column `Total`) of the branches as a dictionary

````python
def salessum(x):
    sum_dict = {}
    sum_A = 0
    sum_B = 0
    sum_C = 0
    for i in x:
        if i[1] == 'A':
            sum_A += float(i[9])
        elif i[1] == 'B':
            sum_B += float(i[9])
        elif i[1] == 'C':
            sum_C += float(i[9])
    sum_dict['A'] = sum_A
    sum_dict['B'] = sum_B
    sum_dict['C'] = sum_C
    return sum_dict

salessum(sales)
````

**a)**  **[5 pts]** Define and implement a function that accepts a tuple of lists and returns the sum of only the unique numbers in that tuple.

For example, if the user enters the following tuple: `
([1, 5], [7, 8, 3, 2], [3, 3, 2])`
then the output should be `26`.

````python
def unique(x):
    x_list = []
    x_unique = []
    sum = 0

    for i in list(x): #converts the input tuple of lists into a simple list
        for j in i:
            x_list.append(j)
    for k in x_list: #creates a list of unique values
        if k not in x_unique:
            x_unique.append(k)
    for l in x_unique: #gets the sum of unique values
        sum = sum + int(l)
    
    return sum

entry_tolist = []
list_totuple = []
tuple_fromlist = ()

while True:
    entry = input("Input a list of numbers, separated by a comma or skip to process:\n")
    if entry == "":
        tuple_fromlist = tuple(list_totuple)
        print("Tuple generated:\t",tuple_fromlist)
        print("Sum of unique values:\t",unique(tuple_fromlist))
        break
    else:
        entry_tolist = entry.split(",")
        list_totuple.append(entry_tolist)
````

For example, if the user enters the following list: 
`
[('Person1', 80), ('Person2', ), ('Person3', 20), ('Person4', 20, 60)]
`

Then the output should be: `{'Person1': 80, 'Person2': None, 'Person3': 20, 'Person4': 20}`


````python
def create_dict(x):
    newdict = {}
    for i in x:
        j = list(i)
        if len(j) == 1: #adds the value None if list only has one element
            j.append(None)
        newdict[j[0]] = j[1]
    return newdict

tuples_list = []

while True:
    entry = str(input("Enter a two-element tuple separated by comma, or skip to process:\n"))
    if entry == "":
        print("Tuple entered:\t",tuples_list)
        print("Dictionary created:\t",create_dict(tuples_list))
        break
    else:
        entry_tolist = entry.split(",")
        entry_totuple = tuple(entry_tolist)
        tuples_list.append(entry_totuple)
````
