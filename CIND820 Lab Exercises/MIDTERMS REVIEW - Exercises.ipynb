Ask the user for a number. Depending on whether the number is even or odd, print out an appropriate message to the user.

```python
while True:
  num = input("Enter an integer, or skip to finish: ")
  if num == "":
    break
  elif int(num) % 2 == 0:
    print("This integer is even.")
  elif int(num) % 2 == 1:
    print("This integer is odd.")

print("End of loop.")
```

Write a program that acquires two integer numbers from a user to calculate and display the
sum and average of the integers in between those two numbers.

```python
x = int(input("Enter first integer: "))
y = int(input("Enter second integer: "))
sum = 0
ave = 0

if x == y:
  print("Error, input integers are the same")
elif x > y:
  for i in range(y+1,x):
    sum += i
    ave = sum / (abs(x - y) - 1)
  print(sum, ave)
elif x < y:
  for i in range(x+1, y):
    sum += i
    ave = sum / (abs(y - x) - 1)
  print(sum, ave)
```

Write the code of a Python program after designing its control-flowchart to read two positive numbers from the user. Those numbers represent the width and length of a two-dimensional shape. The program should display whether the shape is a rectangle or a square based on the equality of the length and width values. Also, design the program, so it prints the area and the perimeter of the shape.

```python
area = 0
perimeter = 0

while True:
  length = int(input("Enter the first positive number: "))
  width = int(input("Enter the second positive number: "))
  if length <= 0 or width <= 0:
    print("Error. Please enter two positive numbers.")
  else:
    area = length * width
    perimeter = 2 * (length + width)
    if length == width:
      print("This shape is a square.","\nArea:", area, "\nPerimeter:", perimeter)
    else:
      print("This shape is a rectangle.","\nArea:", area, "\nPerimeter:", perimeter)
    break
```

Display the total number of: 
question marks, single quotes, and periods within the tedTitles.txt file. 

```python
textFile = open("tedTitles.txt", "r")
content = textFile.readlines()
x = 0
y = 0
z = 0

for i in content:
  for j in i:
    if j == "?":
      x += 1
    if j == "'":
      y += 1
    if j == ".":
      z += 1

print("Question marks: ",x,"\nSingle quotes: ",y,"\nPeriods: ",z)

textFile.close()
```

Display the total number of words that exceed ten letters and less than three letters within the tedTitles.txt file. Hint: Use an outer loop to iterate through the lines of the text and an inner loop to iterate through each line’s words.

```python
textFile = open("tedTitles.txt", "r")
content = textFile.readlines()
count10 = 0
count3 = 0

for i in content:
  for j in i.split():
    if len(j) > 10:
      count10 += 1
    if len(j) < 3:
      count3 += 1

print(count10, count3)

textFile.close()
```
