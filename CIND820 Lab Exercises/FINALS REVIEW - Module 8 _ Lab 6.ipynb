```python
def summation(lower, upper):
  result = 0
  while lower <= upper:
    result += lower
    lower += 1
  return result

summation(1,10)
summation(10,20)
```

```python
class Polygon: #Creating Polygon Class
  width = 0
  height = 0
  def area(self):
    print("The area is: ")
  def perimeter(self):
    print("The perimeter is: ", 2*(self.width + self.height))

class Rectangle(Polygon): #Extending the Polygon Class - INHERITANCE
  def __init__(self, w, h):
    self.width = w
    self.height = h
  def area(self): #Overriding the area() method - POLYMORPHISM
    print("The area is: ", self.width * self.height)
  
class Square(Rectangle): #Extending the Rectangle Class - INHERITANCE
  def __init__(self, side):
    Rectangle.__init__(self, side, side)

Rectangle(10, 20).area()
Rectangle(10, 20).perimeter()
Square(10).area()
Square(10).perimeter()
```
